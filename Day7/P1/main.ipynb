```python
import spacy
from spacy import displacy
import os
import sys
import pandas as pd

nlp = spacy.load("en_core_web_sm")

path = os.path.join(os.getcwd(), "ML471_S7_Datafile_Concept.txt")
with open(path, "r", encoding="utf-8") as f:
    corpus = f.read()

doc_s1 = nlp("The dollar has hit its highest level against the euro after the Federal Reserve head said the US trade deficit is set to stabilize.")
displacy.render(doc_s1, style="dep", jupyter=True, options={"distance": 120})

doc_art = nlp(corpus[:170])
table_data = [[t.text, t.dep_, t.head.text] for t in doc_art if not t.is_space]
table_df = pd.DataFrame(table_data, columns=["Token", "Dependency", "Head Word"])
print(table_df.to_string(index=False))

doc_s3 = nlp("The cat wearing sunglasses danced on the rooftop under the moonlight.")
displacy.render(doc_s3, style="dep", jupyter=True, options={"distance": 120})
```
