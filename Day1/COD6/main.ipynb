```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv('../COD3/ML471_S1_Datafile_Concept.csv')
df.head()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
df['Power_Consumption_diff'] = df[df.columns[0]].diff()
stationary_series = df['Power_Consumption_diff'].dropna()

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

plot_acf(stationary_series, lags=30, ax=axes[0], color='blue')
axes[0].set_title('ACF')
axes[0].set_xlabel('Lags')
axes[0].set_ylabel('ACF Values')
axes[0].set_ylim(-1, 1)

plot_pacf(stationary_series, lags=30, ax=axes[1], color='red')
axes[1].set_title('PACF')
axes[1].set_xlabel('Lags')
axes[1].set_ylabel('PACF Values')
axes[1].set_ylim(-1, 1)

plt.tight_layout()
plt.show()
```
