```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../COD3/ML471_S1_Datafile_Concept.csv')
df.head()
```

```python
diffSeries = df['Consumption'].diff()

plt.figure(figsize=(10, 6))
plt.plot(diffSeries)
plt.title('Differenced Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Differenced Consumption')
plt.legend()    
plt.grid()
```
