```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv('../p2/ML471_S1_Datafile_Practice.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

if 'Date' in df.columns:
    df.set_index('Date', inplace=True)

df_close = df['Close'].resample('ME').mean()

result_add = seasonal_decompose(df_close, model='additive')

fig, axes = plt.subplots(4, 1, figsize=(10, 12))
result_add.observed.plot(ax=axes[0])
axes[0].set_title('Additive Decomposition: Observed')
result_add.trend.plot(ax=axes[1])
axes[1].set_title('Additive Decomposition: Trend')
result_add.seasonal.plot(ax=axes[2])
axes[2].set_title('Additive Decomposition: Seasonal')
result_add.resid.plot(ax=axes[3])
axes[3].set_title('Additive Decomposition: Residual')
plt.tight_layout()
plt.show()

result_mul = seasonal_decompose(df_close, model='multiplicative')

fig, axes = plt.subplots(4, 1, figsize=(10, 12))
result_mul.observed.plot(ax=axes[0])
axes[0].set_title('Multiplicative Decomposition: Observed')
result_mul.trend.plot(ax=axes[1])
axes[1].set_title('Multiplicative Decomposition: Trend')
result_mul.seasonal.plot(ax=axes[2])
axes[2].set_title('Multiplicative Decomposition: Seasonal')
result_mul.resid.plot(ax=axes[3])
axes[3].set_title('Multiplicative Decomposition: Residual')
plt.tight_layout()
plt.show()
```
