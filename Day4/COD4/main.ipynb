```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('../ML471_S4_Datafile_Concept.csv')
df['Datetime'] = pd.to_datetime(df['Datetime'])
df.set_index('Datetime', inplace=True)
model = ExponentialSmoothing(df['Consumption'], trend='add', seasonal='add', seasonal_periods=12).fit()
df['Smoothed'] = model.fittedvalues
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Consumption'], label='Original Data', color='blue')
plt.plot(df.index, df['Smoothed'], label='Final Smoothened Data', color='red')
plt.title('Final Exponential Smoothing')
plt.legend()
plt.show()
```
