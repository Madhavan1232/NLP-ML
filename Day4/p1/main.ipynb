```python
import pandas as pd
import matplotlib.pyplot as plt
import os

df = pd.read_csv(os.path.join('..', 'ML471_S4_Datafile_Practice.csv'))
df['Date'] = pd.to_datetime(df['Date'])
df['Seasonal Naive'] = df['Close'].shift(12)
test_data = df[df['Date'] >= '2014-10-1']

plt.figure(figsize=(12, 8))
plt.plot(test_data['Date'], test_data['Close'], label='Actual')
plt.plot(test_data['Date'], test_data['Seasonal Naive'], label='Seasonal Naive')
plt.xlabel('Date')
plt.ylabel('Closing Stocks')
plt.legend()
plt.show()
```
