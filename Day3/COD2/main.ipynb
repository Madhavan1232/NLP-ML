```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
df = pd.read_csv('../ML471_S2_Datafile_Concept(in).csv')

df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.dropna(subset=['Consumption']).set_index('Datetime')

train_size = int(len(df) * 0.8)
train, test = df['Consumption'][:train_size], df['Consumption'][train_size:]

model = ARIMA(train, order=(2, 1, 2))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(test))

plt.figure(figsize=(10, 6))
plt.plot(train.index, train, label='Train', color='blue')
plt.plot(test.index, test, label='Actual', color='orange', linestyle='--')
plt.plot(test.index, forecast, label='Forecast', color='green', linestyle='--')
plt.title('ARMA Forecast Plot')
plt.legend()
plt.show()
```

```python
residuals = test - forecast

plt.figure(figsize=(10, 4))
plt.plot(test.index, residuals)
plt.title('Residuals')
plt.xlabel('Date')
plt.ylabel('Residuals')
plt.show()
```
