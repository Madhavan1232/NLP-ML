```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
df = pd.read_csv('../ML471_S3_Datafile_Practice.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

df.drop(['SMA_10', 'SMA_30', 'SES'], axis=1, inplace=True)
df.dropna(subset=['Close_diff'], inplace=True)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(df['Close_diff'], ax=axes[0], lags=30)
axes[0].set_title('ACF')
axes[0].set_xlabel('Lags')
axes[0].set_ylabel('ACF Values')

plot_pacf(df['Close_diff'], ax=axes[1], lags=30 , color = 'red')
axes[1].set_title('PACF')
axes[1].set_xlabel('Lags')
axes[1].set_ylabel('PACF Values')

plt.tight_layout()
plt.show()
```
