```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.stats.diagnostic import acorr_ljungbox

df = pd.read_csv('../ML471_S2_Datafile_Concept(in).csv')
df['Datetime'] = pd.to_datetime(df['Datetime'], dayfirst=True)
df.set_index('Datetime', inplace=True)

series = df['Power_Consumption_diff'].dropna()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

plot_acf(series, lags=30, ax=axes[0], color='blue', vlines_kwargs={"colors": 'blue'}, title='ACF')
axes[0].set_xlabel('Lags')
axes[0].set_ylabel('ACF Values')
axes[0].set_ylim(-1, 1)

plot_pacf(series, lags=30, ax=axes[1], color='red', vlines_kwargs={"colors": 'blue'}, title='PACF')
axes[1].set_xlabel('Lags')
axes[1].set_ylabel('PACF Values')
axes[1].set_ylim(-1, 1)

plt.tight_layout()
plt.show()
    
```
